```python
import os

dataset_path = "/content/drive/MyDrive/BreastCancer/CBIS"

# Path to JPEG images
jpeg_folder =  "/content/drive/MyDrive/BreastCancer/CBIS/jpeg"

# Path to CSV files
csv_folder =  "/content/drive/MyDrive/BreastCancer/CBIS/csv"

print("Dataset Path:", dataset_path)
print("JPEG Folder:", jpeg_folder)
print("CSV Folder:", csv_folder)

```

```python
import os

# Check the JPEG folder
print("\n🔍 Checking JPEG Folder:")
jpeg_files = os.listdir("/content/drive/MyDrive/BreastCancer/CBIS/jpeg")
print(f"Total JPEG Images: {len(jpeg_files)}")
print(jpeg_files[:5])  # Show first 5 files

# Check the CSV folder
print("\n🔍 Checking CSV Folder:")
csv_files = os.listdir("/content/drive/MyDrive/BreastCancer/CBIS/csv")
print(f"Total CSV Files: {len(csv_files)}")
print(csv_files)  # Show all CSV files
```

```python
import pandas as pd

# List all Excel files in the CSV folder
csv_files = [file for file in os.listdir(csv_folder) if file.endswith(".xls") or file.endswith(".xlsx")]
dfs = {file: pd.read_excel(os.path.join(csv_folder, file)) for file in csv_files}

for file, df in dfs.items():
    print(f"\n📂 Loaded: {file}")
    print(df.head())  # Show first 5 rows
```

```python

sample_csv = list(dfs.keys())[0]  # Select the first file
df_sample = dfs[sample_csv]

# Generate full image paths
image_paths = df_sample["image file path"].dropna().apply(lambda x: os.path.join(jpeg_folder, x))

# Check if images exist
for img_path in image_paths.head():
    print(f"Exists: {os.path.exists(img_path)} -> {img_path}")
```

```python
import cv2
import matplotlib.pyplot as plt

# Load and display first 5 images
for img_path in image_paths.head():
    if os.path.exists(img_path):
        img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)  # Load in grayscale
        plt.imshow(img, cmap='gray')
        plt.title(os.path.basename(img_path))
        plt.show()
```

```python
import cv2
import numpy as np

def preprocess_image(image_path, target_size=(256, 256)):
    """Load, resize, and normalize an image."""
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)  # Load grayscale image
    img = cv2.resize(img, target_size)  # Resize image
    img = img / 255.0  # Normalize to [0,1]
    return img

# Apply preprocessing to all images
image_data = []
labels = []

for index, row in df_sample.iterrows():
    img_path = os.path.join(jpeg_folder, row["image file path"])
    if os.path.exists(img_path):
        img = preprocess_image(img_path)
        image_data.append(img)
        labels.append(row["pathology"])  # 'pathology' column: BENIGN/MALIGNANT

# Convert to NumPy arrays
image_data = np.array(image_data).reshape(-1, 256, 256, 1)  # Add channel dimension
labels = np.array([1 if label == "MALIGNANT" else 0 for label in labels])  # Convert labels

print(f"Dataset Shape: {image_data.shape}, Labels Shape: {labels.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Split data
X_train, X_test, y_train, y_test = train_test_split(image_data, labels, test_size=0.2, random_state=42)

print(f"Train Set: {X_train.shape}, Test Set: {X_test.shape}")
```
